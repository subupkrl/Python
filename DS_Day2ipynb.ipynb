```python
import numpy as np
high_var_array = [1,100,300,4000,5000]
low_var_array = [2,4,6,8,10]
print(np.std(high_var_array))
print(np.std(low_var_array))
print(np.var(high_var_array))
print(np.var(low_var_array))
print(np.mean(high_var_array))
print(np.mean(low_var_array))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(high_var_array)
plt.show()
plt.hist(low_var_array)
plt.show()
```

```python
#normal distribution
from numpy import random
data = random.normal(loc=1,scale=3,size=(2,3))  #loc-mean, scale-standard deviation
print(data)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(data)
plt.title("Normal Distribution")
plt.show()
```

```python
from numpy import random
data = random.binomial(n=15,p=0.6,size=(2,3))  #loc-mean, scale-standard deviation
print(data)
sns.distplot(data)
plt.title("Binomial Distribution")
plt.show()
```
