<h1> Stemming Words </h1>

```python
from nltk.stem import PorterStemmer
text = 'i am going to the college for meeting and traditional store.'

#tokenization
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
wordtokenize = word_tokenize(text)
print(wordtokenize)

#implementing stemmer - short form ma aauxa testo word ko meaning na huna ni sakxa
ps = PorterStemmer()
for word in wordtokenize:
    print(ps.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
text = 'i am going to the college for meeting and traditional store.'

#tokenization
import nltk
nltk.download('wordnet')
from nltk.tokenize import word_tokenize
wordtokenize = word_tokenize(text)
print(wordtokenize)

#implementing wordnetlemmatizer
lemmatizer = WordNetLemmatizer()
for word in wordtokenize:
    print(lemmatizer.lemmatize(word,pos='v'))
```

```python
# pos_tag ko kaam vaneko part of speech k xa vanera herney ho
from nltk import pos_tag
import nltk
nltk.download('averaged_perceptron_tagger_eng')
from nltk.tokenize import word_tokenize
text = 'i am going to the college for meeting'
tokens = word_tokenize(text)
print(pos_tag(tokens))
```

```python
#bag of words - kunai pani words kati choti repeat vaako xa vanera diney
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
text = np.array([
    'I am going to the college for meeting and traditional store.',
    'I live in Nepal',
    'I study in Ambikeshwari college',
])
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(text)
print(bag_of_words.toarray())
print(vectorizer.vocabulary_)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('punkt_tab')
nltk.download('wordnet')


url = "https://edusanjal.com/"
response = requests.get(url)
html_content = response.content
soup = BeautifulSoup(html_content,'lxml')

first_para = soup.find('p')
print(first_para.text)

wordtokenize = word_tokenize(first_para.text)
print(wordtokenize)

lemmatizer = WordNetLemmatizer()
for word in wordtokenize:
    print(lemmatizer.lemmatize(word))
```

<h1> Handling Date & Time </h1>

```python
import numpy as np
import pandas as pd

date_data = np.array([
    '09-12-2025 02:03:40 PM',
    '05-10-2024 11:40:23 PM',
    '10-08-2025 12:01:40 PM'
])
#convert to datetime
[pd.to_datetime(date,format = '%d-%m-%Y %I:%M:%S %p')for date in date_data]
```

```python
#selecting dates and times
df = pd.DataFrame()
df['Date']=pd.date_range('1/1/2023',periods=150,freq='W')
print(df.head())
print(df.tail())
```

```python
df_date = df[(df['Date'] >= '1/6/2023') & (df['Date'] <= '1/6/2025')]
print(df_date)
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['hour'] = df['Date'].dt.hour
df['minute'] = df['Date'].dt.minute
df.head()
```
