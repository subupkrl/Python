```python
#merge
employee_data = {
    'employee_id':[1,2,3,4],
    'name':['ram','hari','sita','gita']
}
sales_data = {
    'employee_id':[3,4,5,6],
    'total_sales':[23456,2565,675,2450]
}
ed1 = pd.DataFrame(employee_data,columns=['employee_id','name'])
sd1 = pd.DataFrame(sales_data,columns=['employee_id','total_sales'])
pd.merge(ed1,sd1,on='employee_id', how = "outer")  #outer - combination of left and right how = left/right/inner/outer
```

<h1> Handling numerical Dataset </h1>


```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
feature = np.array([[-200.6],[-100.3],[0.0],[100.1],[900.9]])

#create scaler
minmax_scaler = MinMaxScaler(feature_range=(0,1))

#transform the feature
scaled_feature = minmax_scaler.fit_transform(feature)
print(scaled_feature)
print(feature)
```

```python
#StandardScalar
#mean chai 0 and standard deviation chai 1
from sklearn.preprocessing import StandardScaler
feature = np.array([[-200.6],[-100.3],[0.0],[100.1],[900.9]])

standard_scaler = StandardScaler()
scaled_feature = standard_scaler.fit_transform(feature)
print(scaled_feature)
```

```python
print(scaled_feature.mean())
print(scaled_feature.std())
```

<h1> Imputing Missing values </h1>

```python
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
```

```python
#make matrix which include features
feature, _ = make_blobs(n_samples = 1000, n_features = 2, random_state = 1) #random_state ma chai value jun diyeko xa tehi nai hunu paryo change hunu vayena
print(feature)
```

```python
scaler = StandardScaler()
scaled_feature = scaler.fit_transform(feature)
print(scaled_feature)
```

```python
# replace the first features first value with missing value
true_value = scaled_feature[0,0]
scaled_feature[0,0] = np.nan
print(scaled_feature)
```

```python
KNN_imputer = KNNImputer(n_neighbors = 5)  #data should be highly correlated , it is used in small and medium datasets , it is used where imputting is being done from inside
imputer_value = KNN_imputer.fit_transform(scaled_feature)
print(imputer_value)
```

```python
from sklearn.impute import SimpleImputer  #simpleInputer is used for large datasets and it has high speed and data arenot correlated
Simple_imputer = SimpleImputer()
imputer_value = Simple_imputer.fit_transform(scaled_feature)
print(imputer_value)
```
