```python
#pandas - clean,merge,filter,aggregation
#2 data types- series, data frame
#series- 1 dimensional
import pandas as pd
data = [1,2,4,6,7]
series = pd.Series(data,index=('a','b','c','d','e'))
print(series)
```

```python
subject=["IT","DSA","C","C++","Python"]
marks=[50,60,44,56,99]
series = pd.Series(marks,subject)
print(series)
```

```python
print(series.is_unique)  #sabai data unique xa ki nai vanera check garna ko lagi
print(series.index)
print(series.values)
print(series.head(2))
print(series.tail(3))
print(series.sample(3)) # sample means random
print(series.value_counts())
print(series.sort_values())  #for descending bracket ma ascending=False
print(series.sort_index())
print(series.count())
print(series.sum())
print(series.mean())
print(series.median())
print(series.mode())
print(series.std())
print(series.var())
print(series.min())
print(series.max())
print(series.describe())
```
