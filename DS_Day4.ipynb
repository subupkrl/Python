```python
#pivot table
data = {
    'Date' : ['2025-11-01','2025-11-01','2025-11-03','2025-11-03'],
    'Product': ['A','B','A','B'],
    'Sales': [100,200,300,400]
}
print(data)
```

```python
print(pd.pivot_table(d, index='Date', columns='Product', values='Sales',aggfunc='sum'))
```
