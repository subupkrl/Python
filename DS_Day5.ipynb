```python
import pandas as pd
file_path = "/content/drive/MyDrive/country_wise_latest.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
x = df["Confirmed"]
y = df["Deaths"]
plt.scatter(x,y,color='red',marker='+')
x = df["New cases"]
y = df["New deaths"]
plt.title("Scatter plot")
plt.scatter(x,y,color='green')
plt.show()
```

```python
#bargraph = use for bivariate analysis
#numerical vs categorical
x = df["Country/Region"]
y = df["Confirmed"]
plt.figure(figsize=(15,5))  #agadi ko width paxadi ko height
plt.bar(x.head(10),y.head(10))
plt.title("Bar graph")
plt.show()
```

```python
x = df["Country/Region"]
y = df["Deaths"]
plt.figure(figsize=(15,40))  #agadi ko width paxadi ko height
plt.barh(x,y)  #barh gives horizontal bar graph
plt.title("Bar graph")
plt.show()
```

```python
#histogram - univariate analysis , numerical col, usecase-frequency count
data = [35,45,50,10,15,27,61]
plt.hist(data,bins=5) #bins=frequency
plt.title("Histogram plot")
plt.show()
```

```python
data = df["Confirmed"]
plt.hist(data.head(10),bins=10)
plt.show()
```

```python
data = df["Deaths"]
plt.hist(data.head(10),bins=5)
plt.show()
```

```python
#pie chart - univariate/bivariate categorical vs numerical
#use case - to find contribution to a standard scale
country = df["Country/Region"]
data = df["Recovered"]
plt.pie(data.head(5),labels=country.head(5),autopct='%0.2f%%')
plt.title("Pie chart")
plt.show()
```

```python
country = df["Country/Region"]
data = df["Recovered"]
plt.pie(data[12:16],labels=country[12:16],autopct='%0.2f%%') #range for finding mid value
plt.title("Pie chart")
plt.show()
```

```python
subjects = ["Maths","Cloud","GIS","DBMS"]
marks = [60,40,50,90]
plt.pie(marks,labels=subjects,autopct='%0.2f%%',explode=(0.3,0.1,0.2,0.4))
plt.title("Pie chart")
```

```python
import pandas as pd
file_path = "/content/drive/MyDrive/Iris.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df["Species"].value_counts()
```

```python
df["Species"]=df["Species"].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
df['Species'].value_counts()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df["SepalLengthCm"],df["SepalWidthCm"],c=df["Species"],cmap="viridis")
plt.colorbar()
plt.title("Scatter plot")
plt.show()
```

```python
#subplots -> used to represent different plots
ig,ax = plt.subplots(nrows=2,ncols=2,figsize=(14,10))
ax[0,0].scatter(df["PetalLengthCm"],df["PetalWidthCm"],c=df["Species"],cmap="viridis")
ax[0,0].set_xlabel("PetalLengthCm")
ax[0,0].set_ylabel("PetalWidthCm")
ax[0,0].set_title("Petal")

ax[0,1].scatter(df["SepalLengthCm"],df["SepalWidthCm"],c=df["Species"],cmap="viridis")
ax[0,1].set_xlabel("SepalLengthCm")
ax[0,1].set_ylabel("SepalWidthCm")
ax[0,1].set_title("Sepal")

ax[1,0].scatter(df["SepalLengthCm"],df["SepalWidthCm"],c=df["Species"],cmap="viridis")
ax[1,0].set_xlabel("SepalLengthCm")
ax[1,0].set_ylabel("SepalWidthCm")
ax[1,0].set_title("Sepal")

ax[1,1].scatter(df["PetalLengthCm"],df["PetalWidthCm"],c=df["Species"],cmap="viridis")
ax[1,1].set_xlabel("PetalLengthCm")
ax[1,1].set_ylabel("PetalWidthCm")
ax[1,1].set_title("Petal")

plt.show()
```

```python
ig,ax = plt.subplots(nrows=2,ncols=2,figsize=(14,10))
ax[0,0].hist(df["PetalLengthCm"],bins=5)
# ax[0,0].set_xlabel("PetalLengthCm")
# ax[0,0].set_ylabel("PetalWidthCm")
ax[0,0].set_title("Petal")

ax[0,1].hist(df["SepalLengthCm"],bins=5)
# ax[0,1].set_xlabel("SepalLengthCm")
# ax[0,1].set_ylabel("SepalWidthCm")
ax[0,1].set_title("Sepal")

ax[1,0].hist(df["SepalWidthCm"],bins=5)
# ax[1,0].set_xlabel("SepalLengthCm")
# ax[1,0].set_ylabel("SepalWidthCm")
ax[1,0].set_title("Sepal")

ax[1,1].hist(df["PetalWidthCm"],bins=5)
# ax[1,1].set_xlabel("PetalLengthCm")
# ax[1,1].set_ylabel("PetalWidthCm")
ax[1,1].set_title("Petal")

plt.show()
```
