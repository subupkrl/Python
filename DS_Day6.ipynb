```python
import pandas as pd
file_path = "/content/drive/MyDrive/Iris.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df["Species"]=df["Species"].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
df['Species'].value_counts()
```

```python
#3D plot
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8, 10))
ax = plt.subplot(projection = '3d')
ax.scatter(df['SepalLengthCm'], df['SepalWidthCm'], df['PetalLengthCm'],c=df["Species"])
ax.set_title("Iris Dataset Analysis")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")
plt.show()
```

```python
#line plot
x = [0,1,5,30]
y = [0,10,13,0]
z = [0,13,20,9]
fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.plot3D(x, y, z, 'green')
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(projection="3d")
ax.plot3D(df['SepalLengthCm'], df['SepalWidthCm'], df['PetalLengthCm'])
ax.set_title("Line plot")
plt.show()
```

```python
#surface plot
import numpy as np

x = np.linspace(-10,10,100) #linspace- evenely spaced interval
y = np.linspace(-10,10,100)
xx, yy = np.meshgrid(x,y)  #vector bata matrix ma convert garney
z = xx ** 2 + yy ** 2
z.shape
fig = plt.figure(figsize=(8,10))
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,z,cmap="viridis")
ax.plot_surface(xx,yy,z,cmap="viridis")
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap="coolwarm")
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap = "coolwarm")
fig.colorbar(p)
```

```python
#Heatmap
from sklearn.datasets import load_iris  #analysis na garney ML ko model thik xa ki nai vanera check garney
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
print(feature_names)
corr = np.corrcoef(X.T)
fig = plt.figure(figsize=(8,8))
plt.imshow(corr,cmap="coolwarm")
plt.colorbar(label='correlation')
plt.xticks(range(len(feature_names)),feature_names, rotation = 45)
plt.yticks(range(len(feature_names)),feature_names)
plt.title('Correlation Heatmap')
```

```python
import seaborn as sns
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", xticklabels=feature_names, yticklabels=feature_names)
plt.title("Correlation Heatmap")
plt.show()
```
